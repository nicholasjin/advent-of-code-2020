```python
def parse_data(data):
    memory = {}
    for line in data:
        args = line.split(" = ")
#         print(args[0], args[1])
        if args[0] == "mask":
            mask = args[1]
        elif args[0][:3] == "mem":
            register = int(re.findall("\d+", args[0])[0])
#             print(register, args[1])
            memory[register] = apply_mask(args[1], mask)
    return memory
```

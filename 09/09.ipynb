```python
# 
def two_sum(nums, target):
    cache = {}
    for num in nums:
        try:
            cache[target-num]
            return True
        except:
            cache[num] = num
    return False
```

```python
two_sum(data[0:25], data[25])
```

```python
i = 25
while i < len(data):
    if not two_sum(data[i-25:i],data[i]):
        answer_a = data[i]
    i+=1
```

```python
submit(answer_a, part="a", day=9, year=2020)
```
